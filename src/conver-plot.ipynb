```python
import numpy as np
import os
import pandas as pd

import matplotlib.pyplot as plt  
# %matplotlib notebook 
```

## Input

```python
file_path = "../out/"
range_min = 1
range_max = 100
```

## Read files

```python
# read all files in folder
allFiles = os.listdir(file_path)
file_num = len(allFiles)
x = [[] for _ in range(file_num)]
y = [[] for _ in range(file_num)]
labels = []

# print(len(allFiles))

# store the data into x and y
for i, file_name in enumerate(allFiles):
    sub_data = pd.read_table(file_path+file_name, header = None,sep = "\t")
    x[i] = sub_data[0]
    y[i] = sub_data[1]

    
```

```python
# plot the convergency speed under different conditions
fig = plt.figure(figsize=(8, 5))
plt.xlabel('Generation number', fontsize=10)
plt.ylabel('Fitness', fontsize=10)
plt.title('Genetic Algorithms Convergence Speed ', fontsize=10)


for i in range(file_num):
    plt.plot(x[i][range_min:range_max], y[i][range_min:range_max], label = os.path.splitext(allFiles[i])[0])

plt.legend(loc='center left', bbox_to_anchor=(.9, .5))

plt.show()
```
